```python
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True)
results = pose.process(cv2.cvtColor(person_img, cv2.COLOR_BGR2RGB))

if results.pose_landmarks:
    landmarks = results.pose_landmarks.landmark
    # Shoulder and hip landmarks
    left_shoulder = landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER]
    right_shoulder = landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER]
    left_hip = landmarks[mp_pose.PoseLandmark.LEFT_HIP]
    right_hip = landmarks[mp_pose.PoseLandmark.RIGHT_HIP]

    # Convert to pixel coordinates
    def to_pixel(landmark, img):
        h, w = img.shape[:2]
        return int(landmark.x * w), int(landmark.y * h)

    ls = to_pixel(left_shoulder, person_img)
    rs = to_pixel(right_shoulder, person_img)
    lh = to_pixel(left_hip, person_img)
    rh = to_pixel(right_hip, person_img)

    print("Landmarks:", ls, rs, lh, rh)
else:
    print("No pose landmarks detected.")
```
